```python
#Import required libraries

import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
%matplotlib inline
```

```python
#Get current working directory path

os.getcwd()
```

```python
#Getting all file names and concatenating it to single dataframe

files = [file for file in os.listdir('D:\\Data Analytics\\Data Analysis with Python\\Sales_Data\\Dataset')]

final_dataset = pd.DataFrame()

for file in files:
    df = pd.read_csv('D:\\Data Analytics\\Data Analysis with Python\\Sales_Data\\Dataset\\{}'.format(file))
    final_dataset = pd.concat([final_dataset,df])
    
final_dataset.head()
```

# Data Cleaning

```python
#Checking for NA values

final_dataset[final_dataset.isnull().any(axis=1)].shape

#From the shape we can see that there are 545 NA values. But on examining the data, all of them are NA through the row.
#So we can drop this rows.
```

```python
final_dataset.dropna(inplace=True)
final_dataset
```

```python
# Checking duplicate rows

final_dataset[final_dataset.duplicated(keep=False)]
```

```python
# From the dataframe, we could see there are duplicate headers which came into the sheet while importing multiple tables
#Removing this rows

final_dataset.drop_duplicates(inplace=True)
```

```python
#Checking whether Order ID is having any entries with same name

final_dataset.drop(index=final_dataset[final_dataset['Order ID'] == 'Order ID'].index,inplace=True)
```

```python
# Before starting to analyse data, let's see what are the datatypes and convert thme into appropriate one

final_dataset.info()
```

```python
# From the output, 'Quantity Ordered', 'Price Each' and 'Order Date' are string. This can be converted to date and numerics

final_dataset['Quantity Ordered'] = pd.to_numeric(final_dataset['Quantity Ordered'])
final_dataset['Price Each'] = pd.to_numeric(final_dataset['Price Each'])
final_dataset['Order Date'] = pd.to_datetime(final_dataset['Order Date'])

final_dataset.info()
```

# Which was the best month for sales? What was the total sales during this month?

```python
# For this situation we need to extract months from Order Date

final_dataset.head()
```

```python
final_dataset['Month of Order'] = final_dataset['Order Date'].dt.month
final_dataset.tail()
```

```python
final_dataset['Total Sales Amount'] = final_dataset['Quantity Ordered'] * final_dataset['Price Each']
final_dataset.head()
```

```python
sales_by_month = final_dataset.groupby('Month of Order').sum()
sales_by_month
```

```python
plt.figure(figsize=(15,8))

p = sns.barplot(data=sales_by_month,x=sales_by_month.index,y='Total Sales Amount')

#To show full value of y axis labels

plt.ticklabel_format(axis='y',useOffset=False,style='plain')
```

```python
# From the graph, the best month for sale is December and Total sales for this period is 4.6 Million Dollar
```

# Which city has the highest number of sales?

```python
final_dataset.head()
```

```python
final_dataset['City Name'] = final_dataset['Purchase Address'].apply(lambda x : x.split(",")[1]+x.split(",")[2][:3] )
final_dataset.head()
```

```python
sales_by_city = final_dataset.groupby('City Name').sum('Total Sales Amount').sort_values(by='Total Sales Amount',ascending=False)
sales_by_city
```

```python
sns.barplot(data=sales_by_city,x=sales_by_city.index,y='Total Sales Amount')
plt.ticklabel_format(axis='y',useOffset=False,style='plain')
plt.xticks(rotation=90)

plt.show()
```

```python
# From the graph, we can see San Francisco CA got highest sales with Total Sales of 8.25 Million Dollar
```

# Which products are sold together the most?

```python
final_dataset.head()
```

```python
products_in_each_order = final_dataset
```

```python
#Every order will be having a unique Order ID. First let's group by order id and see the products in each order

products_in_each_order['Grouped'] = products_in_each_order.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
products_in_each_order.head()
```

```python
# From the above dataframe , we can see if there is only one occurence of Order ID, it means no other products were sold in that order.
# For easiness, we need only order IDs have more than one occurence

products_in_each_order = products_in_each_order[products_in_each_order['Order ID'].duplicated(keep=False)]

products_in_each_order.head()
```

```python
#Removing entries having same order ID and same Grouoed value

products_in_each_order = products_in_each_order[['Order ID','Grouped']].drop_duplicates()
products_in_each_order.head()
```

```python
# For making different possible combinations, we need to use some external packages

from itertools import combinations
from collections import Counter
```

```python
counter = Counter()

for items in products_in_each_order['Grouped']:
    item = items.split(",")
    counter.update(Counter(combinations(item,2)))
    
for key,value in counter.most_common(10):
    print("{} --> {} sales".format(key,value))
```

```python
# From above analysis 'iPhone with Lightning Charging Cable' and 'Google phone with USB-C Charging Cable' was the best combination
```
