```python
#Importing Libraries

import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Task1: Merge 12 months of data into single csv file

```python
files = [file for file in os.listdir('D:\\Data Analytics\\Data Analysis with Python\\Sales_Data\\Dataset')]

final_dataset = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Dataset/{}".format(file))
    final_dataset = pd.concat([final_dataset,df])

final_dataset
```

```python
final_dataset.head()
```

```python
final_dataset.info()
```

```python
final_dataset.dropna(axis=0,inplace=True)
```

# What was the best month for sales? How much was earned that month?


```python
# Extracting month value from order date

final_dataset['Month'] = final_dataset['Order Date'].str[:2]
```

```python
final_dataset = final_dataset[final_dataset['Month'] != 'Or']
final_dataset.head()
```

```python
final_dataset['Quantity Ordered'] = pd.to_numeric(final_dataset['Quantity Ordered'])

final_dataset['Price Each'] = pd.to_numeric(final_dataset['Price Each'])
```

```python
final_dataset.info()
```

```python
final_dataset['Sales Value'] = final_dataset['Quantity Ordered'] * final_dataset['Price Each']

final_dataset.head()
```

```python
results = final_dataset.groupby('Month').sum()
results
```

```python
plt.figure(figsize=(15,8))

plt.bar(results.index,results['Sales Value'])
```

# Which city had the highest number of sales?

```python
final_dataset['City'] = final_dataset['Purchase Address'].apply(lambda x : x.split(",")[1] + ' '+ x.split(",")[2][:3])

final_dataset.head()
```

```python
city_with_highest_sales = final_dataset.groupby('City').sum()
city_with_highest_sales.sort_values(by='Sales Value',ascending=False)
```

```python
sns.barplot(y=city_with_highest_sales.index,x=city_with_highest_sales['Sales Value'])
```

# Which products are sold together

```python
final_dataset.head()
```

```python
multiple_products_in_single_order_dataset = final_dataset[final_dataset['Order ID'].duplicated(keep=False)]


multiple_products_in_single_order_dataset.head()
```

```python
multiple_products_in_single_order_dataset['Grouped']  = multiple_products_in_single_order_dataset.groupby('Order ID')['Product'].transform(lambda x: ",".join(x))

multiple_products_in_single_order_dataset.head()
```

```python
multiple_products_in_single_order_dataset.head()
```

```python
multiple_products_in_single_order_dataset = multiple_products_in_single_order_dataset[['Order ID','Grouped']].drop_duplicates()

multiple_products_in_single_order_dataset.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in multiple_products_in_single_order_dataset['Grouped']:
    row_list = row.split(",")
    count.update(Counter(combinations(row_list,2)))
    
count.most_common(10)
```

# Which product sold the most

```python
products_sold_most = final_dataset.groupby('Product').sum().sort_values(by='Quantity Ordered',ascending=False)['Quantity Ordered']

products_sold_most = products_sold_most.to_frame()

```

```python
products_sold_most.head()
```

```python
plt.figure(figsize=(25,15))

sns.barplot(data=products_sold_most,y=products_sold_most.index,x='Quantity Ordered')
```
