```python
#Import required libraries

import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
%matplotlib inline
```

```python
#Get current working directory path

os.getcwd()
```

```python
#Getting all file names and concatenating it to single dataframe

files = [file for file in os.listdir('D:\\Data Analytics\\Data Analysis with Python\\Sales_Data\\Dataset')]

final_dataset = pd.DataFrame()

for file in files:
    df = pd.read_csv('D:\\Data Analytics\\Data Analysis with Python\\Sales_Data\\Dataset\\{}'.format(file))
    final_dataset = pd.concat([final_dataset,df])
    
final_dataset.head()
```

# Data Cleaning

```python
#Checking for NA values

final_dataset[final_dataset.isnull().any(axis=1)].shape

#From the shape we can see that there are 545 NA values. But on examining the data, all of them are NA through the row.
#So we can drop this rows.
```

```python
final_dataset.dropna(inplace=True)
final_dataset
```

```python
# Checking duplicate rows

final_dataset[final_dataset.duplicated(keep=False)]
```

```python
# From the dataframe, we could see there are duplicate headers which came into the sheet while importing multiple tables
#Removing this rows

final_dataset.drop_duplicates(inplace=True)
```

```python
#Checking whether Order ID is having any entries with same name

final_dataset.drop(index=final_dataset[final_dataset['Order ID'] == 'Order ID'].index,inplace=True)
```

```python
# Before starting to analyse data, let's see what are the datatypes and convert thme into appropriate one

final_dataset.info()
```

```python
# From the output, 'Quantity Ordered', 'Price Each' and 'Order Date' are string. This can be converted to date and numerics

final_dataset['Quantity Ordered'] = pd.to_numeric(final_dataset['Quantity Ordered'])
final_dataset['Price Each'] = pd.to_numeric(final_dataset['Price Each'])
final_dataset['Order Date'] = pd.to_datetime(final_dataset['Order Date'])

final_dataset.info()
```

# Which was the best month for sales? What was the total sales during this month?

```python
# For this situation we need to extract months from Order Date

final_dataset.head()
```

```python
final_dataset['Month of Order'] = final_dataset['Order Date'].dt.month
final_dataset.tail()
```

```python
final_dataset['Total Sales Amount'] = final_dataset['Quantity Ordered'] * final_dataset['Price Each']
final_dataset.head()
```

```python
sales_by_month = final_dataset.groupby('Month of Order').sum()
sales_by_month
```

```python
plt.figure(figsize=(15,8))

p = sns.barplot(data=sales_by_month,x=sales_by_month.index,y='Total Sales Amount')

#To show full value of y axis labels

plt.ticklabel_format(axis='y',useOffset=False,style='plain')
```

```python
# From the graph, the best month for sale is December and Total sales for this period is 4.6 Million Dollar
```
