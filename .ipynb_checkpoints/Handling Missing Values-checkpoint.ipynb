```python
#Import libraries

import pandas as pd
import numpy as np
```

```python
#Loading dataset

file_location_url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/horse-colic.csv'

dataset=pd.read_csv(file_location_url,na_values='?',header=None)

dataset.head()
```

```python
#Checking missing values

for col in dataset.columns:
    percentage_missing = np.mean(dataset[col].isnull())*100
    print("{} -> {} %".format(col,percentage_missing.round(2)))
    
```
