```python
#Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12,8)

print('Libraries Imported')
```

```python
movies_dataset = pd.read_csv('movies.csv')
movies_dataset.head()
```

```python
#Finding percentage of missing values in each column

for col in movies_dataset.columns:
    percentage_missing = np.mean(movies_dataset[col].isnull())
    print('{} - {}%'.format(col,percentage_missing.round(2)))
```

```python
#Getting more info on Dataset

movies_dataset.info()
```

```python
#Checking Datatypes of columns

movies_dataset.dtypes
```

```python
#Budget and Gross are float.This doesn't make any difference as the values are in millions. So we can convert it into int

movies_dataset['budget'] = movies_dataset['budget'].astype('Int64')
movies_dataset['gross'] = movies_dataset['gross'].astype('Int64')
movies_dataset.head()
```

```python
#Sorting dataset based on Gross amount

movies_dataset.sort_values(by='gross',ascending=False,inplace=True)
```

```python
#Drop any duplicates

movies_dataset.drop_duplicates(inplace=True)
```

```python
movies_dataset.head()
```

```python
#Filling NA values in budget and Gross Revenue with zero

movies_dataset['gross'].fillna( value=0,inplace=True)
movies_dataset['budget'].fillna( value=0,inplace=True)
```

```python
#Scatter plot of Budget vs Gross Amount


plt.scatter(x=movies_dataset['budget'],y=movies_dataset['gross'])
plt.title('Budget vs Gross Amount')


plt.show()
```

```python
movies_dataset['gross'] = movies_dataset['gross'].astype('float64')
movies_dataset['budget'] = movies_dataset['budget'].astype('float64')
```

```python
movies_dataset.dtypes
```

```python
#Plotting using seaborn

sns.regplot(x='budget',y='gross',data =movies_dataset,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
#Let's start looking at correlations
#corr() function only finds correlations between numerical values; ie for this example, it won't include Company Name to correlation table.

movies_dataset.corr()
```

```python
#To make it more easy, we can create a heatmap of correlation with different features

sns.heatmap(movies_dataset.corr(),annot=True)

plt.title('Correlation Heatmap')
```

```python
#As corr() doesn't take into consideration, the string values, we need to convert them into categorical values. 

df_numerized = movies_dataset

for col in df_numerized.columns:
    if (df_numerized[col].dtype == 'object'):
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes
        
df_numerized.head()
```

```python
#Use below code if we need to have a scroll bar for Dataframe view. This is not recommended as it will increase the notebook file size.
#pd.set_option('display.max_rows', None)

corr_matrix = df_numerized.corr()

corr_pairs = corr_matrix.unstack()

sorted_pairs = corr_pairs.sort_values()
```

```python
high_correlation = sorted_pairs[(sorted_pairs > 0.5) & (sorted_pairs < 1)]

high_correlation
```
